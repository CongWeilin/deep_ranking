```python
with open('./data/mslr-web10k/Fold1/test.txt', 'r') as f:
    qid_set = []
    for line in f.readlines():
        qid_num = line.split()[1].split(':')[1]
        qid_set.append(qid_num)
```

```python
qid_set = list(set(qid_set))
train_set = qid_set[:500]
valid_set = qid_set[500:600]
test_set  = qid_set[600:700]
```

```python
train_f = open('./data/mslr-web10k/Fold1/train_mini.txt', 'w')
valid_f = open('./data/mslr-web10k/Fold1/vali_mini.txt', 'w')
test_f  = open('./data/mslr-web10k/Fold1/test_mini.txt', 'w')
f = open('./data/mslr-web10k/Fold1/test.txt', 'r')
for line in f.readlines():
    qid_num = line.split()[1].split(':')[1]
    if qid_num in train_set:
        train_f.write(line)
    elif qid_num in valid_set:
        valid_f.write(line)
    elif qid_num in test_set:
        test_f.write(line)
train_f.close()
valid_f.close()
test_f.close()
f.close()
```
